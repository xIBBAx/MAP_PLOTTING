```python
import plotly.graph_objects as go
import plotly.express as px
```

```python
#SAMPLE
park_names = ['Acadia National Park', 'Arches National Park', 'Badlands National Park', 'Big Bend National Park', 'Biscayne National Park', 'Black Canyon of the Gunnison National Park', 'Bryce Canyon National Park', 'Canyonlands National Park', 'Capitol Reef National Park']
latitudes = [44.35, 38.68, 43.75, 29.25, 25.65, 38.57, 37.57, 38.2, 38.2]
longitudes = [-68.21, -109.57, -102.5, -103.25, -80.08, -107.72, -112.18, -109.93, -111.17]

# SCATTERMAP BOX PLOT
fig = go.Figure(go.Scattermapbox(
    lat=latitudes,
    lon=longitudes,
    mode='markers',
    marker=go.scattermapbox.Marker(
        size=10,
        color='rgb(255, 0, 0)',
        opacity=0.7
    ),
    text=park_names,
))

# MAPBOX API KEY USED HERE
fig.update_layout(
    title='US National Parks',
    autosize=True,
    hovermode='closest',
    mapbox=dict(
        accesstoken='pk.eyJ1IjoiaWJyYWhpbXN1bHRhbiIsImEiOiJjbHUyaWxqZHkwMmt3MmtuNHhtMGJiamM4In0.e6LMfcTyaevLwRYtnF-9Iw',
        bearing=0,
        center=dict(
            lat=37.0902,
            lon=-95.7129
        ),
        pitch=0,
        zoom=3,
        style='light'
    ),
)

fig.show()
```

```python
import folium

m = folium.Map(location=[37.0902, -95.7129], zoom_start=4)

# FOR MARKING PURPOSES
for name, lat, lon in zip(park_names, latitudes, longitudes):
    folium.Marker([lat, lon], popup=name).add_to(m)

# FOR THE PURPOSE OF DISPLAYING THE MAP
m
```
